### Pyplot

Most of the Matplotlib utilities lies under the ** pyplot ** submodule, and are usually imported under the plt alias:



```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Draw a line in a diagram from position (0,0) to position (6,250):

import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
# Plotting Without Line
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

# plt.scatter(xpoints, ypoints) # same
plt.plot(xpoints, ypoints, "o")
plt.show()
```

```python
# Multiple Points
# You can plot as many points as you like, just make sure you have the same number of points in both axis.

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

### Default X-Points

If we do not specify the points in the x-axis, they will get the default values 0, 1, 2, 3, (etc. depending on the length of the y-points.



```python
ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

## Markers

```python
# You can use the keyword argument marker to emphasize each point with a specified marker:

ypoints = np.array([3, 8, 1, 10])

# plt.plot(ypoints,'o')     # different!
plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
# Mark each point with a star:

plt.plot(ypoints, marker = '*')
```

```python
ypoints = np.array([3, 8, 1, 10])


plt.plot(ypoints, 'o:r')    # o defining the style while r is for the color red!
plt.show()
```

```python
# Line Reference(photo)
```

```python
# Marker Size
# plt.plot(ypoints, 'o:g', ms = 20)
plt.plot(ypoints, marker = 'o', ms = 20 )   # ms = val for marker size
plt.show()
```

```python
#Marker Color
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')     # mec for marker edge color
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10]) # why the x-axis plotted as float? 
# You can use the keyword argument markerfacecolor or the shorter mfc 
# to set the color inside the edge of the markers:
plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

```python
# Set the color of both the edge and the face to red:

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r')
plt.show()
```

```python
plt.plot(ypoints, linewidth = '20.5')   # you can also type lw=val
plt.show()
```

## Labels and Title

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)
plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
## Create a Title for a Plot
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
# Set font properties for the title and labels:


x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

```python
# Position the Title
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

## Add Grid Lines to a Plot

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

```python
## Subplots
### With the subplots() function you can draw multiple plots in one figure:
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

#the figure has 1 row, 2 columns, and this plot is the first plot.
plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x2 = np.array([0, 1, 2, 3])
y2 = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)

plt.plot(x2,y2)

plt.show()
```

```python
# with title
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.show()
```

```python
#Super Title
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

## Scatter

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
#Draw two plots on the same figure:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

```python
# color
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#88c999')

plt.show()
```

```python
# color map
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```

```python
# Set your own size for the markers:

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

## Histogram

A histogram is a graph showing frequency distributions.

It is a graph showing the number of observations within each given interval.



```python
#A Normal Data Distribution by NumPy:

x = np.random.normal(170, 10, 250)

print(x)
```

```python
plt.hist(x)
plt.show() 
```

Example: Say you ask for the height of 250 people, you might end up with a histogram like this:

* 2 people from 140 to 145cm
* 5 people from 145 to 150cm
* 15 people from 151 to 156cm
* 31 people from 157 to 162cm
