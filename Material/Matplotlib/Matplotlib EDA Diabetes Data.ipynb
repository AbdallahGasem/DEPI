```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
```

```python
# Load dataset
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target  # Disease progression target

# Display first 5 rows
df.head()
```

# 📊**1. Line Plot: BMI vs. Diabetes Progression**
📌 **Importance:**  
- 📈 Shows how BMI affects diabetes progression.  
- 🔍 Useful for detecting patterns or trends.  

📌 **When to Use:**  
- ✅ When analyzing continuous numerical data (e.g., BMI).  
- ✅ When checking for increasing or decreasing trends.  

📌 **Insights:**  
- If the curve rises, 📊 higher BMI leads to worse diabetes progression.  
- If it's flat, there may be no strong relationship. 

```python
plt.plot(df['bmi'], df['target'], linestyle='-', marker='o', alpha=0.7, color="blue")
plt.xlabel('BMI')
plt.ylabel('Diabetes Progression')
plt.title('📈 BMI vs Diabetes Progression')
plt.show()
```

📌 **When to Use:**

📌 **Insights:**  
- If dots form a rising line, 🔼 high blood pressure may worsen diabetes.

```python
plt.scatter(df['bp'], df['target'], alpha=0.7, color='red')
plt.xlabel('Blood Pressure')
plt.ylabel('Diabetes Progression')
plt.title('Blood Pressure vs Diabetes Progression')
plt.show()
```

# 📊 **3. Histogram: Distribution of BMI Values**


# 🥧 **Pie Chart - Importance, Usage, and Insights**  

## 🎯 **Importance of Pie Charts**  
- 📊 **Visualizes proportions** in a dataset.  
- 🏥 Useful for showing **category distributions** (e.g., gender distribution in diabetes patients).  
- 🎨 **Easy to interpret** when comparing parts of a whole.  

## 🛠 **When to Use a Pie Chart?**  
- ✅ When you need to **compare proportions** of categorical data.  
- 📌 Best for datasets with **limited categories** (too many categories make it hard to read).  
- ⚠️ Avoid when **precise numerical comparison** is required.  

## 🔍 **Insights from Pie Charts (Diabetes Dataset Example)**  
- 🍽️ **Percentage of patients with high BMI vs. normal BMI.**  
- 🔬 **Proportion of males vs. females in the dataset.**  
- 🏥 **Distribution of diabetes risk levels based on cholesterol levels.**  


```python
# Example: Categorizing patients based on BMI (high vs. low)
df['bmi_category'] = np.where(df['bmi'] > df['bmi'].median(), 'High BMI', 'Low BMI')

# Count occurrences
bmi_counts = df['bmi_category'].value_counts()

# Plot Pie Chart
plt.figure(figsize=(4,4))
plt.pie(bmi_counts, labels=bmi_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=30)
plt.title("🥧 BMI Distribution in Diabetes Dataset")
plt.show()
```

## 🛠 **When to Use a Pie Chart?**  
- ✅ When showing **percentage breakdowns** of a categorical feature.  
- 📌 Best for datasets with **a few categories** (2-5 categories).  
- ⚠️ Avoid when you need **precise numeric comparisons**.  

## 🔍 **Insights from Pie Charts (Diabetes Dataset Example - Gender)**  
- 🚻 **Distribution of male vs. female patients** in the dataset.  
- 📊 Helps identify if **one gender is more represented** in diabetes cases.  
- 🔬 Useful in understanding **gender-based risk factors**.

```python
# Categorizing Gender (assuming normalized values around 0 mean male, 1 mean female)
df['gender'] = np.where(df['sex'] > 0, 'Female', 'Male')

# Count occurrences
gender_counts = df['gender'].value_counts()

# Plot Pie Chart
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=90)
plt.title("🚻 Gender Distribution in Diabetes Dataset")
plt.show()
```

📊 **5. Bar Chart: Diabetes Progression by Age Group**
📌 **Importance:**  
- 🏥 Helps compare average diabetes progression across age groups.  
- 📊 Highlights which age groups are at higher risk.  

📌 **When to Use:**  
- ✅ When working with grouped numerical or categorical data.  
- ✅ When comparing trends across categories.  

📌 **Insights:**  
- If older groups have higher bars, 🏥 diabetes worsens with age.  

```python
df['age_group'] = pd.qcut(df['age'], q=4, labels=['Young', 'Middle-Age', 'Older', 'Elderly'])
age_groups = df.groupby('age_group')['target'].mean()

plt.bar(age_groups.index, age_groups.values, color=['green', 'blue', 'orange', 'red'])
plt.xlabel('Age Group')
plt.ylabel('Average Diabetes Progression')
plt.title('📊 Diabetes Progression by Age Group')
plt.show()
```

# 📊 **Horizontal Bar Chart - Importance, Usage, and Insights**  

## 🎯 **Importance of Horizontal Bar Charts**  
- 📊 **Best for comparing categories when labels are long** (e.g., feature importance, risk levels).  
- 📈 Makes it easier to compare **values across multiple categories**.  
- 🚀 Ideal for visualizing **ranked data** in a clean format.  

## 🛠 **When to Use a Horizontal Bar Chart?**  
- ✅ When comparing **group-wise values** with long category names.  
- 📌 Best when working with **ranked or categorical data**.  
- ⚠️ Avoid if the data has a **natural left-to-right progression** (use vertical bars instead).  

## 🔍 **Insights from Horizontal Bar Charts (Diabetes Dataset Example - Age Groups)**  
- 🏥 **Shows which age group has the highest diabetes progression rate**.  
- 📊 **Elderly individuals might show the most significant progression**.  
- 🔬 Useful for **identifying high-risk age groups** for healthcare strategies. 

```python
# Categorizing age into 4 groups
df['age_group'] = pd.qcut(df['age'], q=4, labels=['Young', 'Middle-Age', 'Older', 'Elderly'])

# Calculate the mean progression for each age group
age_groups = df.groupby('age_group')['target'].mean()

# Plot Horizontal Bar Chart
plt.figure(figsize=(4,4))
plt.barh(age_groups.index, age_groups.values, color=['green', 'blue', 'orange', 'red'])
plt.ylabel('🧑 Age Group')
plt.xlabel('📉 Average Diabetes Progression')
plt.title(' Diabetes Progression by Age Group')
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```
