# üìä**1. Line Plot: BMI vs. Diabetes Progression**
üìå **Importance:**  
- üìà Shows how BMI affects diabetes progression.  
- üîç Useful for detecting patterns or trends.  

üìå **When to Use:**  
- ‚úÖ When analyzing continuous numerical data (e.g., BMI).  
- ‚úÖ When checking for increasing or decreasing trends.  

üìå **Insights:**  
- If the curve rises, üìä higher BMI leads to worse diabetes progression.  
- If it's flat, there may be no strong relationship. 

```python
plt.plot(df['bmi'], df['target'], linestyle='-', marker='o', alpha=0.7, color="blue")
plt.xlabel('BMI')
plt.ylabel('Diabetes Progression')
plt.title('üìà BMI vs Diabetes Progression')
plt.show()
```


# ü•ß **Pie Chart - Importance, Usage, and Insights**  

## üéØ **Importance of Pie Charts**  
- üìä **Visualizes proportions** in a dataset.  
- üè• Useful for showing **category distributions** (e.g., gender distribution in diabetes patients).  
- üé® **Easy to interpret** when comparing parts of a whole.  

## üõ† **When to Use a Pie Chart?**  
- ‚úÖ When you need to **compare proportions** of categorical data.  
- üìå Best for datasets with **limited categories** (too many categories make it hard to read).  
- ‚ö†Ô∏è Avoid when **precise numerical comparison** is required.  

## üîç **Insights from Pie Charts (Diabetes Dataset Example)**  
- üçΩÔ∏è **Percentage of patients with high BMI vs. normal BMI.**  
- üî¨ **Proportion of males vs. females in the dataset.**  
- üè• **Distribution of diabetes risk levels based on cholesterol levels.**  


```python
# Example: Categorizing patients based on BMI (high vs. low)
df['bmi_category'] = np.where(df['bmi'] > df['bmi'].median(), 'High BMI', 'Low BMI')

# Count occurrences
bmi_counts = df['bmi_category'].value_counts()

# Plot Pie Chart
plt.figure(figsize=(4,4))
plt.pie(bmi_counts, labels=bmi_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=30)
plt.title("ü•ß BMI Distribution in Diabetes Dataset")
plt.show()
```

## üõ† **When to Use a Pie Chart?**  
- ‚úÖ When showing **percentage breakdowns** of a categorical feature.  
- üìå Best for datasets with **a few categories** (2-5 categories).  
- ‚ö†Ô∏è Avoid when you need **precise numeric comparisons**.  

## üîç **Insights from Pie Charts (Diabetes Dataset Example - Gender)**  
- üöª **Distribution of male vs. female patients** in the dataset.  
- üìä Helps identify if **one gender is more represented** in diabetes cases.  
- üî¨ Useful in understanding **gender-based risk factors**.

```python
# Categorizing Gender (assuming normalized values around 0 mean male, 1 mean female)
df['gender'] = np.where(df['sex'] > 0, 'Female', 'Male')

# Count occurrences
gender_counts = df['gender'].value_counts()

# Plot Pie Chart
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=90)
plt.title("üöª Gender Distribution in Diabetes Dataset")
plt.show()
```

üìä **5. Bar Chart: Diabetes Progression by Age Group**
üìå **Importance:**  
- üè• Helps compare average diabetes progression across age groups.  
- üìä Highlights which age groups are at higher risk.  

üìå **When to Use:**  
- ‚úÖ When working with grouped numerical or categorical data.  
- ‚úÖ When comparing trends across categories.  

üìå **Insights:**  
- If older groups have higher bars, üè• diabetes worsens with age.  

```python
df['age_group'] = pd.qcut(df['age'], q=4, labels=['Young', 'Middle-Age', 'Older', 'Elderly'])
age_groups = df.groupby('age_group')['target'].mean()

plt.bar(age_groups.index, age_groups.values, color=['green', 'blue', 'orange', 'red'])
plt.xlabel('Age Group')
plt.ylabel('Average Diabetes Progression')
plt.title('üìä Diabetes Progression by Age Group')
plt.show()
```

# üìä **Horizontal Bar Chart - Importance, Usage, and Insights**  

## üéØ **Importance of Horizontal Bar Charts**  
- üìä **Best for comparing categories when labels are long** (e.g., feature importance, risk levels).  
- üìà Makes it easier to compare **values across multiple categories**.  
- üöÄ Ideal for visualizing **ranked data** in a clean format.  

## üõ† **When to Use a Horizontal Bar Chart?**  
- ‚úÖ When comparing **group-wise values** with long category names.  
- üìå Best when working with **ranked or categorical data**.  
- ‚ö†Ô∏è Avoid if the data has a **natural left-to-right progression** (use vertical bars instead).  

## üîç **Insights from Horizontal Bar Charts (Diabetes Dataset Example - Age Groups)**  
- üè• **Shows which age group has the highest diabetes progression rate**.  
- üìä **Elderly individuals might show the most significant progression**.  
- üî¨ Useful for **identifying high-risk age groups** for healthcare strategies. 

```python
# Categorizing age into 4 groups
df['age_group'] = pd.qcut(df['age'], q=4, labels=['Young', 'Middle-Age', 'Older', 'Elderly'])

# Calculate the mean progression for each age group
age_groups = df.groupby('age_group')['target'].mean()

# Plot Horizontal Bar Chart
plt.figure(figsize=(4,4))
plt.barh(age_groups.index, age_groups.values, color=['green', 'blue', 'orange', 'red'])
plt.ylabel('üßë Age Group')
plt.xlabel('üìâ Average Diabetes Progression')
plt.title(' Diabetes Progression by Age Group')
plt.gca().invert_yaxis()  # Invert y-axis for better readability
plt.show()
```
