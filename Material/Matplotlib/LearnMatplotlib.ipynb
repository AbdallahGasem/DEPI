```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# Generating random data for applying the visualizations on

x_data = np.random.random(500) * 100
y_data = np.random.random(500) * 100
```

## Scatter Plot

```python
plt.scatter(x_data, y_data)
plt.show()
```

```python
# modify the size, color and the marker style
plt.scatter(x_data, y_data, c="red", s=120, marker="*")
plt.show()
```

## Line Plot

```python
# time sires data
years = [2000 + x for x in range(10)] 
weights = [80,85,110,98,96,92,77,80,81,87]
```

```python
# default plot() plots a line

plt.plot(years, weights)
plt.show()
```

```python
# styling

# plt.plot(years, weights, c="g", lw=12, ls="--")   # -- for the dashed line!
plt.plot(years, weights, "g--", lw= 12)
plt.show()
```

## Bar Chart

```python
# data: poll for voting the best programming language - analysis of categorical data!
langs = ["cpp", "python", "Go", "Java"]
votes = [100, 230, 98, 210]

# ploting the bar chart
plt.bar(langs, votes)
plt.show()
```

```python
# styling
plt.bar(langs, votes, color="red", width=0.3, edgecolor="black")
plt.show()
```

```python
# styling
plt.bar(langs, votes, color="red", width=0.3, edgecolor="black", lw=6)
plt.show()
```

```python
# styling
plt.bar(langs, votes, color="red", width=0.3, edgecolor="black", align="edge")  # the labels are aligned at the edge of each bar
plt.show()
```

## Histogram

```python
# data
data = np.random.normal(20, 1.5, 1000) # generates a normally distrbuted 1000 values with mean = 20 , s= 1.5

plt.hist(data)
plt.show()

```

```python
# styling and Manipulation

plt.hist(data, bins= 20) # specify explicitly the no of pins
plt.show()
```

```python
plt.hist(data, bins=[data.min(), 18, 21, data.max()]) # give a list of ranges for the pins to vary
plt.show()
```

```python
plt.hist(data, bins=20, cumulative=True)    # plot the cummlative Dist function
plt.show()
```

## Pie Chart

```python
plt.pie(votes, labels=langs)
plt.show()
```

```python
# styling and attributes
plt.pie(votes, labels=langs, startangle=90) 
plt.show()
```

```python
# styling and attributes
# plt.pie(votes, labels=langs, autopct="%0.2f%%")     # anotating the % for each
# plt.show()

plt.pie(votes, labels=langs, autopct="%0.2f%%", pctdistance=1.3)     # distancing the % from the center
plt.show()
```

```python
# styling and attributes  

# making some sections sort of float like effect

exploades = [0.1,0,0,0]



plt.pie(votes, labels=langs, explode=exploades, colors=["#4D4DFF", "#6666FF", "#8080FF", "#9999FF"], autopct="%.1f%%")
plt.show()
```

```python
# you can remove any labels from its section then add them through a legend
plt.pie(votes, labels=None)     # distancing the % from the center
plt.legend(labels = langs)
plt.show()
```

## Plot Customization

```python
# labeling the chart

years = np.arange(2022, 2031)
income = [2,2.5,3,3.5,4,4.5,5,5.5,6]
plt.xlabel("Years")     # labeling x-axis
plt.ylabel("Income")    # labeling y-axis
plt.title("Income Vs Years")    # title of the chart
plt.plot(years, income)

plt.yticks(income, [f"{x}k" for x in income])   # used to elaborate what does this line values stands for exactly


plt.show()
```

```python
# You can increase the ticks as follows
# labeling the chart

years = np.arange(2022, 2031)
income = [2,2.5,3,3.5,4,4.5,5,5.5,6]
ytickss = np.arange(1,9)

plt.xlabel("Years")     # labeling x-axis
plt.ylabel("Income")    # labeling y-axis
plt.title("Income Vs Years")    # title of the chart
plt.plot(years, income)

plt.yticks(ytickss, [f"{x}k" for x in ytickss])   # used to elaborate what does this line values stands for exactly


plt.show()
```

## Subploting

```python
## Subplots
### With the subplots() function you can draw multiple plots in one figure:
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

#the figure has 1 row, 2 columns, and this plot is the first plot.
plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x2 = np.array([0, 1, 2, 3])
y2 = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)

plt.plot(x2,y2)

plt.show()
```

```python
# with title
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.show()
```

```python
#Super Title
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

When we use the subplots function in Matplotlib, it returns two objects: the Figure object and an array of Axes objects. Here’s why we specify fig and axs:

fig (short for figure): This represents the entire figure or canvas that contains all the subplots. It allows you to set properties and add overall titles to the figure.

axs (short for axes): This is an array of Axes objects, which represent the individual subplots within the figure. Each Axes object allows you to customize and plot data on the individual subplots.

```python
#another way of doing it 
x = np.arange(1,100)

# This line creates a figure (fig) and an array of subplots (axs) in a 2x2 grid layout.
fig, axs = plt.subplots(2, 2, figsize=(11, 8))    # makes a grid of four locations to carry 4 charts

# access each figure as a 2d array

axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title("Sin Function")

axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title("Cos Function")

axs[1,0].plot(x, np.log(x))
axs[1,0].set_title("log Function")

axs[1,1].plot(x)
axs[1,1].set_title("X dist")

# plt.tight_layout()
fig.suptitle("Testing")
# plt.savefig("fig1.png", dpi=400)    # there are many attributes to be specified here search!
plt.show()
```
